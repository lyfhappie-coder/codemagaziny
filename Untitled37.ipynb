```python
#✅ Multivariate Synthetic Time-Series Dataset Generator (5 Features, 1500 Steps)


import numpy as np
import pandas as pd

# -----------------------------
# CONFIG
# -----------------------------
N = 1500   # total time steps
np.random.seed(42)

# -----------------------------
# BASE TIME INDEX
# -----------------------------
time = np.arange(N)

# -----------------------------
# COMPONENT FUNCTIONS
# -----------------------------
def seasonal_signal(t, period, amplitude=1.0):
    return amplitude * np.sin(2 * np.pi * t / period)

def trend_signal(t, slope=0.001):
    return slope * t

def noise_signal(size, scale=0.2):
    return np.random.normal(scale=scale, size=size)

# -----------------------------
# CREATE MULTIVARIATE FEATURES
# -----------------------------

# Feature 1: Strong daily seasonality + trend + noise
f1 = (
    seasonal_signal(time, period=24, amplitude=1.5) +
    trend_signal(time, slope=0.002) +
    noise_signal(N, scale=0.1)
)

# Feature 2: Weekly seasonality + nonlinear effect
f2 = (
    seasonal_signal(time, period=7*24, amplitude=2.0) +
    0.3 * np.sqrt(time) +
    noise_signal(N, scale=0.15)
)

# Feature 3: Combination of features 1 & 2 + weekly pattern
f3 = (
    0.5 * f1 +
    0.2 * f2 +
    seasonal_signal(time, period=12, amplitude=0.5) +
    noise_signal(N, scale=0.1)
)

# Feature 4: Autoregressive-style series (AR(1)-like)
f4 = np.zeros(N)
for i in range(1, N):
    f4[i] = 0.8 * f4[i-1] + noise_signal(1, scale=0.3)[0]

# Feature 5: Random walk + slight trend
f5 = np.cumsum(noise_signal(N, scale=0.05)) + trend_signal(time, slope=0.001)

# -----------------------------
# CREATE DATAFRAME
# -----------------------------
df = pd.DataFrame({
    "feature_1_daily_seasonal": f1,
    "feature_2_weekly": f2,
    "feature_3_mixed": f3,
    "feature_4_AR_like": f4,
    "feature_5_random_walk": f5,
})

# Add datetime index (optional)
df["timestamp"] = pd.date_range(start="2020-01-01", periods=N, freq="h")
df = df.set_index("timestamp")

# -----------------------------
# DISPLAY RESULT
# -----------------------------
print(df.head())
print("\nShape of dataset:", df.shape)
```

```python
# seq2seq_lstm_attention.py
# PyTorch Seq2Seq LSTM with Bahdanau Attention (autoregressive decoder)
# Requirements: torch, numpy, sklearn (for metrics), tqdm (optional)

import math
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tqdm import tqdm

# ---------------------------
# Config / device
# ---------------------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.manual_seed(42)

# ---------------------------
# Dataset wrapper (works with numpy arrays)
# ---------------------------
class Seq2SeqDataset(Dataset):
    def __init__(self, X: np.ndarray, Y: np.ndarray):
        """
        X: (N, input_len, n_features)
        Y: (N, output_len)  -- single-target per timestep (if multivariate target, change accordingly)
        """
        assert len(X) == len(Y)
        self.X = X.astype(np.float32)
        self.Y = Y.astype(np.float32)
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]


# ---------------------------
# Bahdanau Attention (Additive)
# ---------------------------
class BahdanauAttention(nn.Module):
    def __init__(self, enc_hidden_dim: int, dec_hidden_dim: int):
        super().__init__()
        # project encoder and decoder hidden states into common space
        self.W_enc = nn.Linear(enc_hidden_dim, enc_hidden_dim, bias=False)
        self.W_dec = nn.Linear(dec_hidden_dim, enc_hidden_dim, bias=False)
        self.v = nn.Linear(enc_hidden_dim, 1, bias=False)

    def forward(self, encoder_outputs: torch.Tensor, decoder_hidden: torch.Tensor):
        """
        encoder_outputs: (batch, seq_len, enc_hidden)
        decoder_hidden:  (batch, dec_hidden)
        returns:
           context: (batch, enc_hidden)
           attn_weights: (batch, seq_len)
        """
        # (batch, seq_len, enc_hidden)
        enc_t = self.W_enc(encoder_outputs)
        # (batch, 1, enc_hidden)
        dec_t = self.W_dec(decoder_hidden).unsqueeze(1)
        # broadcast add -> (batch, seq_len, enc_hidden)
        score = torch.tanh(enc_t + dec_t)
        # (batch, seq_len, 1) -> (batch, seq_len)
        e = self.v(score).squeeze(-1)
        attn_weights = torch.softmax(e, dim=1)
        # context: weighted sum (batch, enc_hidden)
        context = torch.bmm(attn_weights.unsqueeze(1), encoder_outputs).squeeze(1)
        return context, attn_weights


# ---------------------------
# Seq2Seq model: Encoder + Attention + Decoder
# ---------------------------
class Seq2SeqLSTMAttn(nn.Module):
    def __init__(self,
                 input_size: int,
                 enc_hidden_size: int = 128,
                 dec_hidden_size: int = 128,
                 enc_num_layers: int = 1,
                 dec_num_layers: int = 1,
                 dropout: float = 0.1,
                 output_size: int = 1):
        """
        input_size: number of input features per timestep
        output_size: dimension of predicted value per timestep (usually 1, the target)
        """
        super().__init__()
        self.enc_hidden_size = enc_hidden_size
        self.dec_hidden_size = dec_hidden_size
        self.enc_num_layers = enc_num_layers
        self.dec_num_layers = dec_num_layers
        self.output_size = output_size

        # encoder LSTM
        self.encoder = nn.LSTM(input_size, enc_hidden_size,
                               num_layers=enc_num_layers, batch_first=True,
                               dropout=dropout if enc_num_layers > 1 else 0.0, bidirectional=False)

        # decoder (we'll feed previous output as input each step)
        # decoder input is previous predicted value (output_size) optionally concatenated with context
        # we'll feed only the previous output; the attention context is merged inside decoder step
        # use an LSTMCell for step-by-step decoding
        self.decoder_cell = nn.LSTMCell(output_size + enc_hidden_size, dec_hidden_size)

        # attention
        self.attention = BahdanauAttention(enc_hidden_size, dec_hidden_size)

        # output projection from decoder hidden state -> prediction
        self.out_proj = nn.Linear(dec_hidden_size, output_size)

    def forward(self, encoder_inputs: torch.Tensor, decoder_targets: torch.Tensor = None,
                teacher_forcing_ratio: float = 0.5):
        """
        encoder_inputs: (batch, in_len, input_size)
        decoder_targets: (batch, out_len) or None (for inference). Values should be unscaled/normalized as used in training.
        Returns:
          outputs: (batch, out_len, output_size)
          attn_weights_seq: (batch, out_len, in_len)  # attention per decoder step
        """
        batch_size = encoder_inputs.size(0)
        in_len = encoder_inputs.size(1)

        # --- Encode ---
        encoder_outputs, (h_n, c_n) = self.encoder(encoder_inputs)  # encoder_outputs: (batch, in_len, enc_hidden)
        # initialize decoder hidden state from encoder final hidden state (project if different sizes)
        # take last layer hidden state (h_n: (num_layers, batch, hidden))
        dec_h = torch.zeros(batch_size, self.dec_hidden_size, device=encoder_inputs.device)
        dec_c = torch.zeros(batch_size, self.dec_hidden_size, device=encoder_inputs.device)
        # Option: initialize decoder hidden by a linear projection of encoder last hidden
        # if enc_hidden_size != dec_hidden_size, project:
        if self.enc_hidden_size != self.dec_hidden_size:
            # simple projection via linear
            proj_h = nn.Linear(self.enc_hidden_size, self.dec_hidden_size).to(encoder_inputs.device)
            proj_c = nn.Linear(self.enc_hidden_size, self.dec_hidden_size).to(encoder_inputs.device)
            # use the last layer
            dec_h = proj_h(h_n[-1])
            dec_c = proj_c(c_n[-1])
        else:
            dec_h = h_n[-1]
            dec_c = c_n[-1]

        # prepare first decoder input: we'll use the last observed target from encoder_inputs if available
        # assume the target is the first feature in encoder_inputs (common pattern). If you have separate last target,
        # you should pass it instead.
        # Here, just take encoder_inputs[:, -1, 0:output_size]
        prev_output = encoder_inputs[:, -1, 0:self.output_size]  # (batch, output_size)
        outputs = []
        attn_weights_seq = []

        out_len = decoder_targets.size(1) if decoder_targets is not None else 0
        max_decode_steps = out_len if decoder_targets is not None else 24  # default inference horizon if not provided

        # ---- decode step-by-step ----
        for t in range(max_decode_steps):
            # attention context computed from encoder outputs and current decoder hidden
            context, attn_weights = self.attention(encoder_outputs, dec_h)  # context (batch, enc_hidden), attn_weights (batch, in_len)

            # decoder input is concat(prev_output, context)
            dec_input = torch.cat([prev_output, context], dim=1)  # (batch, output_size + enc_hidden)
            dec_h, dec_c = self.decoder_cell(dec_input, (dec_h, dec_c))  # each: (batch, dec_hidden)

            # produce output
            out_step = self.out_proj(dec_h)  # (batch, output_size)
            outputs.append(out_step.unsqueeze(1))
            attn_weights_seq.append(attn_weights.unsqueeze(1))  # (batch,1,in_len)

            # decide whether to use teacher forcing
            if decoder_targets is not None and (torch.rand(1).item() < teacher_forcing_ratio):
                # use true target at t
                prev_output = decoder_targets[:, t].unsqueeze(1)  # (batch,1) -> if output_size>1 adjust slicing
            else:
                prev_output = out_step  # model prediction

        outputs = torch.cat(outputs, dim=1)  # (batch, out_len, output_size)
        attn_weights_seq = torch.cat(attn_weights_seq, dim=1)  # (batch, out_len, in_len)
        return outputs, attn_weights_seq


# ---------------------------
# Training & evaluation helpers
# ---------------------------
def rmse(y_true, y_pred):
    return math.sqrt(mean_squared_error(y_true.flatten(), y_pred.flatten()))

def train_model(model, train_loader, val_loader=None, epochs=20, lr=1e-3,
                teacher_forcing_ratio=0.5, save_path=None):
    model = model.to(DEVICE)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    criterion = nn.MSELoss()
    best_val = float("inf")

    for epoch in range(1, epochs + 1):
        model.train()
        train_losses = []
        for xb, yb in train_loader:
            xb = xb.to(DEVICE)  # (batch, in_len, n_features)
            yb = yb.to(DEVICE)  # (batch, out_len)
            optimizer.zero_grad()
            preds, _ = model(xb, decoder_targets=yb, teacher_forcing_ratio=teacher_forcing_ratio)
            # preds: (batch, out_len, 1) -> squeeze last dim
            preds_squeezed = preds.squeeze(-1)
            loss = criterion(preds_squeezed, yb)
            loss.backward()
            optimizer.step()
            train_losses.append(loss.item())
        avg_train = float(np.mean(train_losses))

        # validation
        if val_loader is not None:
            model.eval()
            val_losses = []
            with torch.no_grad():
                for xb, yb in val_loader:
                    xb = xb.to(DEVICE)
                    yb = yb.to(DEVICE)
                    preds, _ = model(xb, decoder_targets=yb, teacher_forcing_ratio=0.0)  # no teacher forcing in val
                    preds_squeezed = preds.squeeze(-1)
                    loss = criterion(preds_squeezed, yb)
                    val_losses.append(loss.item())
            avg_val = float(np.mean(val_losses))
            print(f"Epoch {epoch:02d} - Train MSE: {avg_train:.6f}  Val MSE: {avg_val:.6f}")
            if avg_val < best_val:
                best_val = avg_val
                if save_path:
                    torch.save(model.state_dict(), save_path)
        else:
            print(f"Epoch {epoch:02d} - Train MSE: {avg_train:.6f}")

    # load best model if saved
    if save_path:
        model.load_state_dict(torch.load(save_path, map_location=DEVICE))
    return model


def evaluate_model(model, data_loader):
    model.eval()
    preds_list = []
    y_list = []
    attn_list = []
    with torch.no_grad():
        for xb, yb in data_loader:
            xb = xb.to(DEVICE)
            yb = yb.to(DEVICE)
            preds, attn = model(xb, decoder_targets=None, teacher_forcing_ratio=0.0)  # autoregressive inference
            preds = preds.squeeze(-1).cpu().numpy()  # (batch, out_len)
            yb = yb.cpu().numpy()
            preds_list.append(preds)
            y_list.append(yb)
            attn_list.append(attn.cpu().numpy())  # (batch, out_len, in_len)
    preds = np.vstack(preds_list)
    y = np.vstack(y_list)
    attn = np.vstack(attn_list)
    return preds, y, attn


# ---------------------------
# Example usage (connect with preprocessed arrays)
# ---------------------------
if __name__ == "__main__":
    # --- placeholders: replace these with your actual arrays from preprocessing ---
    # X_train shape: (N_train, input_len, n_features)
    # Y_train shape: (N_train, output_len)
    # X_val, Y_val, X_test, Y_test likewise
    #
    # For demonstration we create tiny random arrays. Replace them!
    input_len = 48
    output_len = 24
    n_features = 5
    N_train = 800
    N_val = 150
    N_test = 150

    # Replace these lines with your preprocessed arrays
    X_train = np.random.randn(N_train, input_len, n_features).astype(np.float32)
    Y_train = np.random.randn(N_train, output_len).astype(np.float32)
    X_val = np.random.randn(N_val, input_len, n_features).astype(np.float32)
    Y_val = np.random.randn(N_val, output_len).astype(np.float32)
    X_test = np.random.randn(N_test, input_len, n_features).astype(np.float32)
    Y_test = np.random.randn(N_test, output_len).astype(np.float32)

    # Create DataLoaders
    batch_size = 64
    train_loader = DataLoader(Seq2SeqDataset(X_train, Y_train), batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(Seq2SeqDataset(X_val, Y_val), batch_size=batch_size, shuffle=False)
    test_loader = DataLoader(Seq2SeqDataset(X_test, Y_test), batch_size=batch_size, shuffle=False)

    # Build model (choose sizes to match your problem)
    model = Seq2SeqLSTMAttn(input_size=n_features,
                            enc_hidden_size=128,
                            dec_hidden_size=128,
                            enc_num_layers=1,
                            dec_num_layers=1,
                            dropout=0.1,
                            output_size=1)

    # Train
    model = train_model(model, train_loader, val_loader, epochs=10, lr=1e-3,
                        teacher_forcing_ratio=0.5, save_path="best_seq2seq_attn.pth")

    # Evaluate / infer
    preds, y_true, attn_weights = evaluate_model(model, test_loader)
    print("Preds shape:", preds.shape, "True shape:", y_true.shape, "Attn shape:", attn_weights.shape)

    # Example metrics
    mse = ((preds - y_true)**2).mean()
    rmse_val = math.sqrt(mse)
    mae_val = np.mean(np.abs(preds - y_true))
    print(f"Test RMSE: {rmse_val:.6f}, MAE: {mae_val:.6f}")

    # Example: average attention for first decoder timestep
    avg_attn_first_step = attn_weights[:, 0, :].mean(axis=0)  # (in_len,)
    print("avg attention for decoder step 0 (shape):", avg_attn_first_step.shape)
```

```python
import os, math, time, warnings
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

import torch, torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm

warnings.filterwarnings("ignore")
torch.manual_seed(42); np.random.seed(42)

# ----------------
# 1) Generate data
# ----------------
def generate_multivariate_series(n_steps=1500, seed=42):
    rng = np.random.RandomState(seed)
    t = np.arange(n_steps)
    trend = 0.0005 * (t ** 1.2)
    daily = 10 * np.sin(2 * np.pi * t / 24)
    weekly = 5 * np.sin(2 * np.pi * t / (24 * 7))
    long = 3 * np.sin(2 * np.pi * t / (24 * 365 + 30))
    exog1 = 0.8 * daily + 0.2 * rng.normal(scale=2, size=n_steps)
    exog2 = 2.0 * np.cos(2 * np.pi * t / 12) + rng.normal(scale=1.5, size=n_steps)
    exog3 = rng.normal(scale=3.0, size=n_steps)
    base_target = 50 + trend + daily + weekly * 0.6 + long * 0.4
    noise_scale = 1.0 + 0.05 * np.abs(base_target)
    noise = rng.normal(scale=noise_scale, size=n_steps)
    target = base_target + 0.5 * exog1 + 0.2 * exog2 + noise
    df = pd.DataFrame({
        "target": target,
        "exog1": exog1,
        "exog2": exog2,
        "exog3": exog3,
        "exog4": 0.1*base_target + rng.normal(scale=0.5, size=n_steps)
    })
    df.index = pd.date_range("2000-01-01", periods=n_steps, freq="H")
    return df

df = generate_multivariate_series(1500)
print("Dataset shape:", df.shape)

# ----------------
# 2) Preprocess (scaling + windowing)
# ----------------
INPUT_WINDOW = 48
OUTPUT_WINDOW = 24
TEST_RATIO = 0.15
VAL_RATIO = 0.15

n = len(df)
test_size = int(n * TEST_RATIO)
val_size = int(n * VAL_RATIO)
train_size = n - val_size - test_size

df_train = df.iloc[:train_size]
df_val = df.iloc[train_size:train_size+val_size]
df_test = df.iloc[train_size+val_size:]

scaler = StandardScaler().fit(df_train.values)
train_scaled = scaler.transform(df_train.values)
val_scaled = scaler.transform(df_val.values)
test_scaled = scaler.transform(df_test.values)

def create_windows(arr, in_w, out_w):
    X, Y = [], []
    N = len(arr)
    for i in range(N - in_w - out_w + 1):
        X.append(arr[i:i+in_w])
        Y.append(arr[i+in_w:i+in_w+out_w, 0])  # target is column 0
    return np.asarray(X), np.asarray(Y)

X_train, Y_train = create_windows(train_scaled, INPUT_WINDOW, OUTPUT_WINDOW)
X_val, Y_val = create_windows(val_scaled, INPUT_WINDOW, OUTPUT_WINDOW)
X_test, Y_test = create_windows(test_scaled, INPUT_WINDOW, OUTPUT_WINDOW)

print("Shapes: X_train, Y_train:", X_train.shape, Y_train.shape)
print("Shapes: X_val, Y_val:", X_val.shape, Y_val.shape)
print("Shapes: X_test, Y_test:", X_test.shape, Y_test.shape)

# ----------------
# 3) Model definitions (Bahdanau attention)
# ----------------
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class Seq2SeqDataset(Dataset):
    def __init__(self,X,Y):
        self.X = X.astype(np.float32); self.Y = Y.astype(np.float32)
    def __len__(self): return len(self.X)
    def __getitem__(self,idx): return self.X[idx], self.Y[idx]

class BahdanauAttention(nn.Module):
    def __init__(self, enc_h, dec_h):
        super().__init__()
        self.W_enc = nn.Linear(enc_h, enc_h, bias=False)
        self.W_dec = nn.Linear(dec_h, enc_h, bias=False)
        self.v = nn.Linear(enc_h,1,bias=False)
    def forward(self, enc_out, dec_h):
        enc_t = self.W_enc(enc_out)
        dec_t = self.W_dec(dec_h).unsqueeze(1)
        score = torch.tanh(enc_t + dec_t)
        e = self.v(score).squeeze(-1)
        weights = torch.softmax(e, dim=1)
        context = torch.bmm(weights.unsqueeze(1), enc_out).squeeze(1)
        return context, weights

class Seq2SeqLSTMAttn(nn.Module):
    def __init__(self,input_size,enc_h=64,dec_h=64,enc_layers=1,output_size=1,dropout=0.1):
        super().__init__()
        self.enc = nn.LSTM(input_size, enc_h, num_layers=enc_layers, batch_first=True, dropout=dropout if enc_layers>1 else 0.0)
        self.attn = BahdanauAttention(enc_h, dec_h)
        self.dec_cell = nn.LSTMCell(output_size + enc_h, dec_h)
        self.out = nn.Linear(dec_h, output_size)
        self.enc_h=enc_h; self.dec_h=dec_h
    def forward(self, encoder_inputs, decoder_targets=None, tf_ratio=0.5):
        b = encoder_inputs.size(0)
        enc_out, (h_n, c_n) = self.enc(encoder_inputs)
        if self.enc_h != self.dec_h:
            proj_h = nn.Linear(self.enc_h, self.dec_h).to(encoder_inputs.device)
            proj_c = nn.Linear(self.enc_h, self.dec_h).to(encoder_inputs.device)
            dec_h = proj_h(h_n[-1]); dec_c = proj_c(c_n[-1])
        else:
            dec_h = h_n[-1]; dec_c = c_n[-1]
        prev = encoder_inputs[:, -1, 0:1]
        outputs=[]; attns=[]
        max_steps = decoder_targets.shape[1] if decoder_targets is not None else OUTPUT_WINDOW
        for t in range(max_steps):
            context, weights = self.attn(enc_out, dec_h)
            dec_in = torch.cat([prev, context], dim=1)
            dec_h, dec_c = self.dec_cell(dec_in, (dec_h, dec_c))
            out_step = self.out(dec_h)
            outputs.append(out_step.unsqueeze(1))
            attns.append(weights.unsqueeze(1))
            if decoder_targets is not None and (torch.rand(1).item() < tf_ratio):
                prev = decoder_targets[:, t].unsqueeze(1)
            else:
                prev = out_step
        outputs = torch.cat(outputs, dim=1)
        attns = torch.cat(attns, dim=1)
        return outputs, attns

# ----------------
# 4) Training & small HPO grid
# ----------------
def train_one(model, train_loader, val_loader, epochs=6, lr=1e-3, tf_ratio=0.5, weight_decay=1e-6):
    model = model.to(DEVICE)
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, factor=0.5, patience=2)
    crit = nn.MSELoss()
    best_val = 1e9; best_state=None
    for ep in range(1, epochs+1):
        model.train()
        for xb, yb in train_loader:
            xb = xb.to(DEVICE); yb = yb.to(DEVICE)
            opt.zero_grad()
            preds, _ = model(xb, decoder_targets=yb, tf_ratio=tf_ratio)
            loss = crit(preds.squeeze(-1), yb)
            loss.backward(); opt.step()
        # validation
        val_loss = 0.0; cnt=0
        model.eval()
        with torch.no_grad():
            for xb,yb in val_loader:
                xb=xb.to(DEVICE); yb=yb.to(DEVICE)
                preds,_ = model(xb, decoder_targets=yb, tf_ratio=0.0)
                val_loss += crit(preds.squeeze(-1), yb).item() * xb.size(0); cnt += xb.size(0)
        val_loss /= (cnt+1e-12)
        scheduler.step(val_loss)
        if val_loss < best_val:
            best_val = val_loss
            best_state = {k:v.cpu() for k,v in model.state_dict().items()}
    model.load_state_dict(best_state)
    return model, best_val

# HPO grid (small)
grid = {"lr":[1e-3, 3e-4], "batch":[64,128], "enc_h":[64,128]}
results=[]
for lr in grid["lr"]:
    for batch in grid["batch"]:
        for enc_h in grid["enc_h"]:
            print(f"Config lr={lr}, batch={batch}, enc_h={enc_h}")
            train_loader = DataLoader(Seq2SeqDataset(X_train, Y_train), batch_size=batch, shuffle=True, num_workers=2, pin_memory=torch.cuda.is_available())
            val_loader = DataLoader(Seq2SeqDataset(X_val, Y_val), batch_size=batch, shuffle=False, num_workers=2, pin_memory=torch.cuda.is_available())
            model = Seq2SeqLSTMAttn(input_size=X_train.shape[2], enc_h=enc_h, dec_h=enc_h, output_size=1)
            t0=time.time()
            model_trained, val_loss = train_one(model, train_loader, val_loader, epochs=6, lr=lr, tf_ratio=0.5)
            t1=time.time()
            print(f"  Val loss={val_loss:.6f} time={t1-t0:.1f}s")
            results.append({"lr":lr,"batch":batch,"enc_h":enc_h,"val_loss":val_loss,"time_s":t1-t0,"state":model_trained.state_dict()})

res_df = pd.DataFrame([{"lr":r["lr"], "batch":r["batch"], "enc_h":r["enc_h"], "val_loss":r["val_loss"], "time_s":r["time_s"]} for r in results])
print("\nHPO results (top 5):\n", res_df.sort_values("val_loss").head())

# pick best
best = min(results, key=lambda r: r["val_loss"])
print("Best config:", best["lr"], best["batch"], best["enc_h"], "val_loss", best["val_loss"])

# ----------------
# 5) Retrain best on train+val, evaluate on test
# ----------------
X_trainval = np.vstack([X_train, X_val]); Y_trainval = np.vstack([Y_train, Y_val])
trainval_loader = DataLoader(Seq2SeqDataset(X_trainval, Y_trainval), batch_size=best["batch"], shuffle=True, num_workers=2, pin_memory=torch.cuda.is_available())
test_loader = DataLoader(Seq2SeqDataset(X_test, Y_test), batch_size=best["batch"], shuffle=False, num_workers=2, pin_memory=torch.cuda.is_available())

best_model = Seq2SeqLSTMAttn(input_size=X_train.shape[2], enc_h=best["enc_h"], dec_h=best["enc_h"], output_size=1)
best_model.load_state_dict(best["state"])
best_model, _ = train_one(best_model, trainval_loader, test_loader, epochs=4, lr=best["lr"], tf_ratio=0.3)

def eval_model(model, loader):
    model.eval()
    preds=[]; ys=[]; atts=[]
    with torch.no_grad():
        for xb,yb in loader:
            xb=xb.to(DEVICE); yb=yb.to(DEVICE)
            p,a = model(xb, decoder_targets=None, tf_ratio=0.0)
            preds.append(p.squeeze(-1).cpu().numpy()); ys.append(yb.cpu().numpy()); atts.append(a.cpu().numpy())
    return np.vstack(preds), np.vstack(ys), np.vstack(atts)

preds_model, y_test_true, attns = eval_model(best_model, test_loader)

def unscale_target(arr_scaled, scaler_obj):
    if arr_scaled.size==0: return arr_scaled
    n,h = arr_scaled.shape
    flat = arr_scaled.reshape(-1)
    tmp = np.zeros((len(flat), scaler_obj.mean_.shape[0]))
    tmp[:,0] = flat
    inv = scaler_obj.inverse_transform(tmp)[:,0]
    return inv.reshape(n,h)

preds_model_unscaled = unscale_target(preds_model, scaler)
y_test_unscaled = unscale_target(y_test_true, scaler)

def metrics(y_true, y_pred):
    rmse = math.sqrt(mean_squared_error(y_true.flatten(), y_pred.flatten()))
    mae = mean_absolute_error(y_true.flatten(), y_pred.flatten())
    mape = np.mean(np.abs((y_true - y_pred) / np.clip(np.abs(y_true),1e-6,None))) * 100.0
    true_sum = np.sum(y_true, axis=1); pred_sum = np.sum(y_pred, axis=1)
    da = np.mean(np.sign(true_sum[1:]-true_sum[:-1]) == np.sign(pred_sum[1:]-pred_sum[:-1])) if len(true_sum)>1 else float("nan")
    return {"RMSE":rmse,"MAE":mae,"MAPE(%)":mape,"DirectionalAcc":da}

model_metrics = metrics(y_test_unscaled, preds_model_unscaled)
print("Seq2Seq test metrics (unscaled):", model_metrics)

# ----------------
# 6) Baselines (rolling origin) - ARIMA & ExponentialSmoothing
# ----------------
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing

orig_train = df_train['target'].values
orig_val = df_val['target'].values
orig_test = df_test['target'].values
full_series = np.concatenate([orig_train, orig_val, orig_test])
test_global_start = train_size + val_size
n_test_windows = X_test.shape[0]
h = OUTPUT_WINDOW

arima_preds = []
es_preds = []
for i in tqdm(range(n_test_windows), desc="Baselines rolling"):
    origin = test_global_start + i
    history = full_series[:origin]
    # ARIMA small order for speed
    try:
        arima = ARIMA(history, order=(2,0,1)).fit(method="innovations_mle", disp=0)
        f_arima = arima.forecast(steps=h)
    except Exception:
        last = history[-1]
        f_arima = np.array([last]*h)
    arima_preds.append(f_arima)
    try:
        es = ExponentialSmoothing(history, trend="add", seasonal=None).fit()
        f_es = es.forecast(steps=h)
    except Exception:
        last = history[-1]
        f_es = np.array([last]*h)
    es_preds.append(f_es)

arima_preds = np.vstack(arima_preds)
es_preds = np.vstack(es_preds)

# ground-truth windows for test
true_windows = []
for i in range(n_test_windows):
    idx = test_global_start + i
    true_windows.append(full_series[idx:idx+h])
true_windows = np.vstack(true_windows)

arima_metrics = metrics(true_windows, arima_preds)
es_metrics = metrics(true_windows, es_preds)
model_metrics_vs_true = metrics(true_windows, preds_model_unscaled)

print("\nFinal metrics (all on original scale):")
print("Seq2Seq-LSTM-Attn:", model_metrics_vs_true)
print("ARIMA baseline:", arima_metrics)
print("ExpSmoothing baseline:", es_metrics)

# Summarize and plot few example forecast windows
res_df = pd.DataFrame({
    "method":["Seq2Seq-LSTM-Attn","ARIMA","ExpSmoothing"],
    "RMSE":[model_metrics_vs_true["RMSE"], arima_metrics["RMSE"], es_metrics["RMSE"]],
    "MAE":[model_metrics_vs_true["MAE"], arima_metrics["MAE"], es_metrics["MAE"]],
    "MAPE(%)":[model_metrics_vs_true["MAPE(%)"], arima_metrics["MAPE(%)"], es_metrics["MAPE(%)"]],
    "DirectionalAcc":[model_metrics_vs_true["DirectionalAcc"], arima_metrics["DirectionalAcc"], es_metrics["DirectionalAcc"]]
})
print("\nSummary:\n", res_df)

for w in range(min(3, n_test_windows)):
    plt.figure(figsize=(8,3))
    plt.plot(true_windows[w], label="True")
    plt.plot(preds_model_unscaled[w], label="Seq2Seq")
    plt.plot(arima_preds[w], label="ARIMA")
    plt.plot(es_preds[w], label="ExpSmooth")
    plt.legend(); plt.title(f"Window {w}"); plt.xlabel("Step ahead"); plt.show()

# Save summary CSV
res_df.to_csv("forecast_comparison_summary.csv", index=False)
print("Saved forecast_comparison_summary.csv")
```
